### Aníkúlápó (Movie) Data Creation

```python
# Importing Libries

from urllib.request import urlopen
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

```python
# Setting display options

pd.set_option('display.width', 30000)
pd.set_option('display.max_columns', 1000)
pd.set_option('display.max_colwidth', 1000)
```

Loading the web page

```python
# Requesting the web page

r = requests.get('https://en.wikipedia.org/wiki/An%C3%ADk%C3%BAl%C3%A1p%C3%B3')
```

```python
# converting it to a beautiful soup object

bs_object = bs(r.content)
```

```python
# To know the content of the beautiful soup, it will print out the HTML of the page

content = bs_object.prettify()
print(content)
```

```python
# Extracting the needed info

bs_object.find(class_ = 'infobox vevent')
```

```python
# Assigning a variable to the needed info class

info_box = bs_object.find(class_ = 'infobox vevent')
print(info_box.prettify())
```

```python
# To extract the table rows of your interested info

info_box.find_all('tr')

# Rename the variable
 
info_rows = info_box.find_all('tr')
```

```python
# Using FOR LOOP to iterate the rows

for row in info_rows:
    print(row.prettify())
```

```python
# defining a function to find all the list in the rows

def get_content_value(row):
    if row.find('li'):
        return str(li.get_text() for li in row.find_all('li'))
    else:
        return row.find('td').get_text()
```

```python
movie_info = {}

for index, row in enumerate(info_rows):
    if index == 0:
        movie_info['Title'] = row.find('th').get_text()
        
    elif index == 1:
        continue
        
    else:
        content_key = row.find('th').get_text()
        content_value = get_content_value(row)
        movie_info[content_key]= content_value
        
movie_info
```

```python
# cleaning up the extracted data

def get_content_value(row):
    if row.find('li'):
        return str([li.get_text().replace('\xa0', ' ').replace('[1]', '') for li in row.find_all('li')])
    else:
        return row.find('td').get_text().replace('\xa0', ' ').replace('[1]', '')
```

```python
cleaned_movie_info = {}

for index, row in enumerate(info_rows):
    if index == 0:
        cleaned_movie_info['Title'] = row.find('th').get_text(' ', strip=True)
        
    elif index == 1:
        continue
        
    else:
        content_key = row.find('th').get_text(' ', strip=True)
        content_value = get_content_value(row)
        cleaned_movie_info[content_key]= content_value
        
cleaned_movie_info
```

```python
# Converting the extracteddict to a df

movie = pd.DataFrame(list(cleaned_movie_info.items()))
movie
```
